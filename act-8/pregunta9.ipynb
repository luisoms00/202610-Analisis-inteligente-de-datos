```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Creación del Dataset con los valores de la imagen
data = {
    'Dia': list(range(1, 26)),
    'Ingreso': [
        1452, 1361, 1426, 1470, 1456, 1430, 1354, 1442, 1394, 1459, 
        1399, 1458, 1537, 1425, 1445, 1439, 1348, 1450, 1431, 1446, 
        1485, 1405, 1461, 1490, 1426
    ],
    'Habitaciones_Ocupadas': [
        23, 47, 21, 39, 37, 29, 23, 44, 45, 16, 
        30, 42, 54, 27, 34, 15, 19, 38, 44, 47, 
        43, 38, 51, 61, 39
    ]
}

df = pd.DataFrame(data)

# Generar la Gráfica de Correlación Lineal (Dispersión con línea de regresión)
plt.figure(figsize=(10, 6))
sns.regplot(x='Habitaciones_Ocupadas', y='Ingreso', data=df, ci=None, line_kws={"color": "red"})

plt.title('Relación entre Habitaciones Ocupadas e Ingreso del Restaurante')
plt.xlabel('Habitaciones Ocupadas (Variable Independiente)')
plt.ylabel('Ingreso del Restaurante (Variable Dependiente)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Cálculo del Coeficiente de Correlación de Pearson
correlacion = df['Habitaciones_Ocupadas'].corr(df['Ingreso'])

print(f"Coeficiente de Correlación de Pearson (r): {correlacion:.4f}")

# Interpretación automática básica
if abs(correlacion) < 0.3:
    print("Interpretación: Correlación débil o inexistente.")
elif 0.3 <= abs(correlacion) < 0.7:
    print("Interpretación: Correlación moderada.")
else:
    print("Interpretación: Correlación fuerte.")
```
