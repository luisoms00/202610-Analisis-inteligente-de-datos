```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
X_train = diabetes_X[:-20]
X_test = diabetes_X[-20:]
y_train = diabetes_y[:-20]
y_test = diabetes_y[-20:]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```

```python
# Graficar
plt.figure(figsize=(8, 6))


plt.scatter(X_test, y_test, color='black', alpha=0.5, label='Puntos') 


plt.plot(X_test, regr.predict(X_test), color='blue', linewidth=3, label='Modelo Lineal') 

plt.title('Regresión Lineal: BMI vs Target')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Progresión de Diabetes')
plt.legend()
plt.show()
```

# Conclusion
### Aunque existe una correlación clara y positiva (a mayor BMI, mayor tendencia a la progresión de la enfermedad), el BMI no cuenta la historia completa, el 53% restante se puede deber a otros factores que el modelo no toma en cuenta.
