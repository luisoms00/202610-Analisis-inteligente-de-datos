```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

def get_prices(symbol, start, end):
    data = yf.download(symbol, start=start, end=end)
    return data["Open"].values
```

```python
prices = get_prices("MSFT", "2010-01-01", "2024-01-01")
```

```python
plt.plot(prices)
plt.title("Opening stock prices")
plt.xlabel("days")
plt.ylabel("price")
plt.show()
```
